```python
import numpy as np
import pandas as pd 
import sklearn as sk
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
```

```python
train_df.drop(columns=['Fence','MiscFeature','FireplaceQu','Alley','PoolQC','SalePrice'])
train_df = train_df.fillna(train_df.mean())
```

```python
y = train_df['SalePrice'].to_numpy()
x = train_df.iloc[:,1:80]
x = pd.get_dummies(x).to_numpy()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=45)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train, y_train)
prediction = reg.predict(x_test)
prediction_error = (y_test - prediction)**2
prediction_error.sum()/y_test.shape[0]
```

The ordinary least square worked ok here with MSE:534332841 and R square 0.93. 

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
b = []
for k in range(1,30):
    a = []
    kf = KFold(n_splits=10)
    for train, test in kf.split(x, y):
        neigh = KNeighborsRegressor(n_neighbors=k)
        neigh.fit(x[train], y[train]) 
        prediction = neigh.predict(x[test])
        prediction_error = (y[test] - prediction)**2
        a.append(prediction_error.sum()/y[test].shape[0])
    b.append(a)
```

```python
s = []
for each in b:
    s.append(np.mean(each))
std = []
for each in b:
    std.append(np.std(each))
plt.errorbar(range(1,30), s, std, linestyle='None', marker='^')
plt.show()
```

The KNN here with k = 2 is bad with MSE=2215597661. Again the knn do not work well in high dimensional data sets.

```python
import statsmodels.api as sm
y = train_df['SalePrice']
x = train_df.iloc[:,1:80]
x = pd.get_dummies(x)
```

```python
y = train_df['SalePrice']
x = train_df.iloc[:,1:80]
x = pd.get_dummies(x)
def BSF(x_train,y_train, n):
    included = set(x_train.columns)
    while len(included) > n:
        X2 = sm.add_constant(x_train[included])
        BSR = sm.OLS(y_train, X2)
        result = BSR.fit()
        k = result.pvalues
        try:
            k1 = k.drop('const')
        except:
            k1 = k
        included.remove(argmax(k1))
    return(included)
```

```python
def test_and_score_(Inc,x_train,y_train,x_test,y_test):
    X2 = sm.add_constant(x_train[Inc])
    BSR = sm.OLS(y_train, X2)
    result = BSR.fit()
    X3 = sm.add_constant(x_test[Inc])
    pre_result = result.predict(X3)
    prediction_error = (y_test-pre_result)**2
    return (prediction_error.sum()/y_test.shape[0])
```

```python
y = train_df['SalePrice']
x = train_df.iloc[:,1:80]
x = pd.get_dummies(x)
MSE = []
for a in range(5,10):
    MSE_one_group = 0
    for train, test in kf.split(x, y):
        x_train = x.iloc[train]
        y_train = y.iloc[train]
        x_test  = x.iloc[test]
        y_test  = y.iloc[test]
        w = BSF(x_train,y_train,a)
        score = test_and_score_(w,x_train,y_train,x_test,y_test)
        MSE_one_group = MSE_one_group + score
    MSE.append(MSE_one_group/10)
print(argmin(MSE))
plt.plot(range(5,80),MSE)
```

For the Backward stepwise (linear) regression , we choose k = 7 from the 10 fold validation with MSE= 1,588,279,908. Surpirsingly the backward stepwise here did less well. Simply eliminate features will lose information thus yeild less promising result.

```python
df_corr = pd.get_dummies(train_df)
columns_corr = df_corr.corr()['SalePrice'].abs()
l = columns_corr.sort_values()
columns_list = np.array(l.index)
columns_list[-6:-1]
```

```python
MSE = []
for a in range(5,80):
    for train, test in kf.split(x, y):
        MSE_one_group = 0
        x_train = x.iloc[train]
        y_train = y.iloc[train]
        x_test  = x.iloc[test]
        y_test  = y.iloc[test]
        number_of_features = - (a + 1)
        w = columns_list[number_of_features:-1]
        score = test_and_score_(w,x_train,y_train,x_test,y_test)
        MSE_one_group = MSE_one_group + score
    MSE.append(MSE_one_group/10)
print(argmin(MSE))
plt.plot(range(5,80),MSE)
```
